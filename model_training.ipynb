```python
df_edited = df.groupby('smoker').filter(lambda x : len(x)>4)
df_edited.info()
df_edited['smoker'].value_counts()
```

```python
df_edited['smoker'] = df_edited['smoker'].astype('category')
```

```python
df_edited['smoker Encoded'] = df_edited['smoker'].cat.codes
df_edited.head()
```

```python
df_edited.dtypes
```

```python
df_edited['region'] = df_edited['region'].astype('category')
```

```python
df_edited.dtypes
```

```python
df_edited['region Encoded'] = df_edited['region'].cat.codes
df_edited.head()
```

```python
df_edited['sex'] = df_edited['sex'].astype('category')
```

```python
df_edited['sex Encoded'] = df_edited['sex'].cat.codes
df_edited.head()
```

```python
df_final = df_edited.drop(['sex', 'region', 'smoker'], axis=1)
df_final.head()
```

```python
df_final.info()
```

```python
import numpy as np
import pickle
import sklearn

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
X = df_final.iloc[:,[0,1,3,4,5]]
y = df_final.iloc[:,2]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
err = np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
y_pred, y_test
```

```python
err
```

```python
print(pd.DataFrame(zip(X.columns, reg.coef_)))
print(reg.intercept_)
```

```python
pickle.dump(reg, open('model.pkl', 'wb'))
```
